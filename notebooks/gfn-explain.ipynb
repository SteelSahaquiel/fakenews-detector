```python
import numpy as np
import spacy
from lime.lime_text import LimeTextExplainer

nlp = spacy.load('../model')
```

```python
def predict_proba(text):
    #lst = list(nlp(text).cats.values())
    lst = [list(nlp(s).cats.values()) for s in text] 
    lst = np.asarray(lst)
    lst = lst.round(decimals=3)
    print(lst)
    return lst
```

```python
text = "ΑΠΟΦΑΣΗ-ΓΡΟΘΙΑ ΑΠΟ ΤΡΕΙΣ ΔΙΚΑΣΤΕΣ!"
```

```python
explainer = LimeTextExplainer(class_names=['REAL', 'FAKE'])

exp = explainer.explain_instance(text, predict_proba, num_samples=200, num_features=10)
```
