```python
import numpy as np
import spacy
#from lime.lime_text import LimeTextExplainer
from eli5.lime import TextExplainer

nlp = spacy.load('../model')
```

```python
def predict_proba(text):
    #lst = list(nlp(text).cats.values())
    lst = [list(nlp(s).cats.values()) for s in text] 
    lst = np.asarray(lst)
    lst = lst.round(decimals=3)
    lst[-1] = [1,0]
    return lst
```

```python
text = "Από την μια, βλέπουν Μητροπολίτες και Ιερείς να προτρέπουν τον κόσμο έμμεσα ή άμεσα με το παράδειγμά τους ώστε να κάνει το εμβόλιο κι από την άλλη κυρίως γνωστοί και φωτισμένοι γέροντες του Αγίου Όρος, να παρακινούν να αποφύγουν τον εμβολιασμό με κάθε τρόπο. Ο λαός του Θεού έχει διχασθεί, στους υποστηρικτές του εμβολίου και τους αρνητές του, εν μέσω μιας τελείως συγκεχυμένης κατάστασης, με συνεχόμενες αντιφατικές πληροφορίες, οι οποίες από την μια μας λένε ότι εάν εμβολιαστείς θα έχεις ανοσία για κάποιους μήνες κι από την άλλη ότι πολλοί, με ασθενείς οργανισμούς, μόλις κάνανε το εμβόλιο πεθάνανε σε λίγες ημέρες (πρόσφατα στη Νορβηγία)..."
```

```python
te = TextExplainer(n_samples = 100)
te.fit(text, predict_proba)
te.show_prediction()
```
